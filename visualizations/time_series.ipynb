```python
import pandas as pd
import matplotlib.pyplot as plt

DRUG_FILES = ['US_Amphetamine.csv', 'US_Codeine.csv', 'US_Hydrocodone.csv', 'US_Morphine.csv', 'US_Oxycodone.csv']
DATE_COLUMN_NAME = ['Date:RetailDrugDistribution_DrugDistribution_Amphetamine', 'Date:RetailDrugDistribution_DrugDistribution_Codeine', 'Date:RetailDrugDistribution_DrugDistribution_Hydrocodone', 'Date:RetailDrugDistribution_DrugDistribution_Morphine', 'Date:RetailDrugDistribution_DrugDistribution_Oxycodone']
VALUE_COLUMN_NAME = ['Value:RetailDrugDistribution_DrugDistribution_Amphetamine', 'Value:RetailDrugDistribution_DrugDistribution_Codeine', 'Value:RetailDrugDistribution_DrugDistribution_Hydrocodone', 'Value:RetailDrugDistribution_DrugDistribution_Morphine', 'Value:RetailDrugDistribution_DrugDistribution_Oxycodone']
```

```python
for drug_i in range(len(DRUG_FILES)):
    df = pd.read_csv('./data/drugs/' + DRUG_FILES[drug_i])

    # sum number of cases for all counties at a given time 
    df = df.groupby(DATE_COLUMN_NAME[drug_i], as_index=False)[VALUE_COLUMN_NAME[drug_i]].sum()

    X = df[DATE_COLUMN_NAME[drug_i]]
    Y = df[VALUE_COLUMN_NAME[drug_i]]
    
    plt.plot(X, Y, label=DRUG_FILES[drug_i][3:-4])
    plt.xlabel('Date')
    plt.ylabel('Number of cases')
    plt.title(f'Retail {DRUG_FILES[drug_i][3:-4]} Distribution Cases Over Time')
    plt.xticks(ticks=range(0, len(X), 4), labels=X[::4], rotation=90)
    
plt.legend()
plt.show()
```

```python
# Colorado drug use over time graph
for drug_i in range(len(DRUG_FILES)):
    df = pd.read_csv('./data/drugs/' + DRUG_FILES[drug_i])

    # includes only counties in Colorado
    df = df[df['placeDcid'].str.contains('geoId/08')]
    
    # sum number of cases for all counties at a given time 
    df = df.groupby(DATE_COLUMN_NAME[drug_i], as_index=False)[VALUE_COLUMN_NAME[drug_i]].sum()

    X = df[DATE_COLUMN_NAME[drug_i]]
    Y = df[VALUE_COLUMN_NAME[drug_i]]

    plt.plot(X, Y, label=DRUG_FILES[drug_i][3:-4])
    plt.xlabel('Date')
    plt.ylabel('Number of cases')
    plt.title(f'Retail {DRUG_FILES[drug_i][3:-4]} Distribution Cases Over Time In Colorado')
    plt.xticks(ticks=range(0, len(X), 4), labels=X[::4], rotation=90)
    
plt.legend()
plt.axvline(x='2012-10', color='red', linestyle='--', linewidth=2)
plt.show()
```
