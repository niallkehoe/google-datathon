```python
import pandas as pd
import matplotlib.pyplot as plt

DRUG_FILES = ['US_Amphetamine.csv', 'US_Codeine.csv', 'US_Hydrocodone.csv', 'US_Morphine.csv', 'US_Oxycodone.csv']
DATE_COLUMN_NAME = ['Date:RetailDrugDistribution_DrugDistribution_Amphetamine', 'Date:RetailDrugDistribution_DrugDistribution_Codeine', 'Date:RetailDrugDistribution_DrugDistribution_Hydrocodone', 'Date:RetailDrugDistribution_DrugDistribution_Morphine', 'Date:RetailDrugDistribution_DrugDistribution_Oxycodone']
VALUE_COLUMN_NAME = ['Value:RetailDrugDistribution_DrugDistribution_Amphetamine', 'Value:RetailDrugDistribution_DrugDistribution_Codeine', 'Value:RetailDrugDistribution_DrugDistribution_Hydrocodone', 'Value:RetailDrugDistribution_DrugDistribution_Morphine', 'Value:RetailDrugDistribution_DrugDistribution_Oxycodone']
```

```python
for drug_i in range(len(DRUG_FILES)):
    df = pd.read_csv('../data/drugs/' + DRUG_FILES[drug_i])

    # sum number of cases for all counties at a given time 
    df = df.groupby(DATE_COLUMN_NAME[drug_i], as_index=False)[VALUE_COLUMN_NAME[drug_i]].sum()

    X = df[DATE_COLUMN_NAME[drug_i]]
    Y = df[VALUE_COLUMN_NAME[drug_i]]
    
    plt.plot(X, Y, label=DRUG_FILES[drug_i][3:-4])
    plt.xlabel('Date')
    plt.ylabel('Number of cases')
    plt.title(f'Retail {DRUG_FILES[drug_i][3:-4]} Distribution Cases Over Time')
    plt.xticks(ticks=range(0, len(X), 4), labels=X[::4], rotation=90)
    
plt.legend()
plt.show()
```

```python
# drug use over time graph for illegal states
for drug_i in range(len(DRUG_FILES)):
    df = pd.read_csv('../data/drugs/' + DRUG_FILES[drug_i])

    # removes Colorado, Oregon, Washington, Alaska, and Nevada from data
    df = df[~df['placeDcid'].str.contains('geoId/08')]
    df = df[~df['placeDcid'].str.contains('geoId/41')]
    df = df[~df['placeDcid'].str.contains('geoId/53')]
    df = df[~df['placeDcid'].str.contains('geoId/02')]
    df = df[~df['placeDcid'].str.contains('geoId/32')]
    
    # sum number of cases for all counties at a given time 
    df = df.groupby(DATE_COLUMN_NAME[drug_i], as_index=False)[VALUE_COLUMN_NAME[drug_i]].sum()

    X = df[DATE_COLUMN_NAME[drug_i]]
    Y = df[VALUE_COLUMN_NAME[drug_i]]

    plt.plot(X, Y, label=DRUG_FILES[drug_i][3:-4])
    plt.xlabel('Date')
    plt.ylabel('Number of cases')
    plt.title(f'Retail {DRUG_FILES[drug_i][3:-4]} Distribution Cases Over Time In Illegal States')
    plt.xticks(ticks=range(0, len(X), 4), labels=X[::4], rotation=90)
    
plt.legend()
plt.show()
```

```python
# percentage of legal state cases to all state cases
for drug_i in range(len(DRUG_FILES)):
    df_all_states = pd.read_csv('../data/drugs/' + DRUG_FILES[drug_i])

    substrings = ['geoId/08', 'geoId/41', 'geoId/53', 'geoId/02', 'geoId/32']
    pattern = '|'.join(substrings)
    
    # selects only Colorado, Oregon, Washington, Alaska, and Nevada from data
    df_legal_states = df_all_states[df_all_states['placeDcid'].str.contains(pattern)]
    
    # sum number of cases for all counties at a given time 
    df_legal_states = df_legal_states.groupby(DATE_COLUMN_NAME[drug_i], as_index=False)[VALUE_COLUMN_NAME[drug_i]].sum()
    df_all_states = df_all_states.groupby(DATE_COLUMN_NAME[drug_i], as_index=False)[VALUE_COLUMN_NAME[drug_i]].sum()

    X = df_all_states[DATE_COLUMN_NAME[drug_i]]
    Y = df_legal_states[VALUE_COLUMN_NAME[drug_i]] / df_all_states[VALUE_COLUMN_NAME[drug_i]]

    plt.plot(X, Y, label=DRUG_FILES[drug_i][3:-4])
    plt.xlabel('Date')
    plt.ylabel('Number of cases')
    plt.title(f'Share Of Retail Drug Distribution For Legal States Over Time')
    plt.xticks(ticks=range(0, len(X), 4), labels=X[::4], rotation=90)
    
plt.legend()
plt.show()
```

```python
import plotly.graph_objects as go

# Create an empty figure
fig = go.Figure()

# Texas drug use over time graph
for drug_i in range(len(DRUG_FILES)):
    # Read the CSV file
    df = pd.read_csv('../data/drugs/' + DRUG_FILES[drug_i])

    # Include only counties in Texas
    df = df[df['placeDcid'].str.contains('geoId/48')]

    # Sum the number of cases for all counties at a given time
    df = df.groupby(DATE_COLUMN_NAME[drug_i], as_index=False)[VALUE_COLUMN_NAME[drug_i]].sum()

    X = df[DATE_COLUMN_NAME[drug_i]]
    Y = df[VALUE_COLUMN_NAME[drug_i]]

    # Add a trace for the current drug data
    fig.add_trace(go.Scatter(
        x=X,
        y=Y,
        mode='lines',
        name=DRUG_FILES[drug_i][3:-4]
    ))

# Update layout properties
fig.update_layout(
    title='Retail Drug Distribution In mg Over Time In Texas',
    xaxis_title='Date',
    yaxis_title='Total mg',
    xaxis=dict(
        tickmode='array',
        tickvals=X[::4],
        ticktext=[str(date) for date in X[::4]],
        tickangle=45
    ),
    legend_title='Drug Types'
)

# Show the plot
fig.show()
```
