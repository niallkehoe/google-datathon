```python
import pandas as pd
import matplotlib.pyplot as plt

DRUG_FILES = ['US_Amphetamine.csv', 'US_Codeine.csv', 'US_Hydrocodone.csv', 'US_Morphine.csv', 'US_Oxycodone.csv']
DATE_COLUMN_NAME = ['', 'Date:RetailDrugDistribution_DrugDistribution_Codeine', '', '', '']
VALUE_COLUMN_NAME = ['', 'Value:RetailDrugDistribution_DrugDistribution_Codeine', '', '', '']
```

```python
df = pd.read_csv('./data/drugs/' + DRUG_FILES[1])

# sum number of cases for all counties at a given time 
df = df.groupby(DATE_COLUMN_NAME[1], as_index=False)[VALUE_COLUMN_NAME[1]].sum()

X = df['Date:RetailDrugDistribution_DrugDistribution_Codeine']
Y = df['Value:RetailDrugDistribution_DrugDistribution_Codeine']
```

```python
plt.plot(X, Y)
plt.xlabel('Date')
plt.ylabel('Number of cases')
plt.title('Retail Codeine Drug Distribution Cases Over Time')
plt.xticks(ticks=range(0, len(X), 4), labels=X[::4], rotation=90)
plt.show()
```
